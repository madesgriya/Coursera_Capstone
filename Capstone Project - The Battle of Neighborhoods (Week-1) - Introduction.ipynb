```python
import folium
from geopy.geocoders import Nominatim

address = 'Kuta, Bali'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Kuta are {}, {}.'.format(latitude, longitude))

map_bali = folium.Map(location=[latitude, longitude], zoom_start=10.5)
map_bali
```
